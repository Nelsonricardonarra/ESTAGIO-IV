```python
# Importar as bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Ler os dados
B_Dados = pd.read_csv('/content/Unicorns in april 2022s end.csv')

# Verificar Dimensão
B_Dados.shape
```

```python
# Primeiros registros
B_Dados.head()
```

```python
# Colunas
B_Dados.columns
```

```python
# Renomear colunas
B_Dados.columns=['Companhia', 'Valor ($B)', 'Data de Adesão', 'País', 'Cidade ',
       'Industria', 'Investidores']
```

```python
display(B_Dados)
```

```python
# Verificar Tipo do informação
B_Dados.info()
```

```python
# Convertendo a data que eatá como objeto

B_Dados['Data de Adesão']= pd.to_datetime(B_Dados['Data de Adesão'])
B_Dados['Data de Adesão']
```

```python
# Transformando a coluna Valor
B_Dados['Valor ($B)'] = pd.to_numeric( B_Dados['Valor ($B)'].apply( lambda Linha: Linha.replace('$', '') ) )


```

```python
# Extrair o Ano e Mes da data de adesão

date_col= pd.DatetimeIndex( B_Dados['Data de Adesão'] )

B_Dados['Ano'] =date_col.year
B_Dados['Mes'] =date_col.month

```

```python
# Visualizando após criar duas colunas " Ano" e "Mês"
# Visualizando a coluna de Valor em Bilhões sem o cifrão

B_Dados.head()
```

```python
# Campos nulos
B_Dados.isnull().sum()
```

```python
# Analisando campos nulos

plt.figure( figsize=(15,6) )
plt.title('Analisando Campos Nulos')
sns.heatmap( B_Dados.isnull(), cbar=False );


```

```python
# Retirando os campos nulos
B_Dados=B_Dados.dropna()
```

```python
# Visualizando a ausência dos campos nulos
plt.figure( figsize=(8,6) )
plt.title('Campos Nulos Zerados')
sns.heatmap(B_Dados.isnull(),cmap='plasma',cbar=False)
plt.show
```
